```python
import json

def read_json(filename):
    with open(filename) as f:
        data = f.read()
        data = json.loads(data)
    return data

filename = "IMDB_movies_merged.json"

data = read_json(filename)
```

```python
import pandas as pd
df = pd.json_normalize(data)
df
```

```python

#* Q1 What are the director names of the top five rating films of the actor Harrison Ford (who had been casted in 'Indiana Jones and the Raiders of the Lost Ark') that have not been directed by Steven Spielberg?
def q1():
    top_five = []
    # get data
    for film in data:
        try:
            rating = float(film["ratingValue"])
            cast_list = film["cast"]
            cast_name_list = [actor["name"] for actor in cast_list]
            director = film["director"]
        except (KeyError, ValueError):
            continue

        if "Harrison Ford" in cast_name_list and director["name"] != "Steven Spielberg":
            top_five.append((director["name"], rating))

    top_five = sorted(top_five, key=lambda x: x[1], reverse=True)
    top_five_rating = set()
    i = 0
    while len(top_five_rating) < 5:
        top_five_rating.add(top_five[i][1])
        i += 1

    top_five = [i[0] for i in sorted(top_five, key=lambda x: x[0]) if i[1] in top_five_rating]
    for a in top_five:
        print(a)

#* Q2 What is the top rated film (not directed either by George Lucas or by Steven Spielberg) in which Harrison Ford and Tommy Lee Jones casting together?
def q2():
    top_five = []
    for film in data:
        try:
            name = film["name"]
            rating = float(film["ratingValue"])
            cast_list = film["cast"]
            cast_name_list = [actor["name"] for actor in cast_list]
            director = film["director"]
        except (KeyError, ValueError):
            continue

        if director not in ("George Lucas", "Steven Spielberg") and "Harrison Ford" in cast_name_list and "Tommy Lee Jones" in cast_name_list:
            top_five.append((name, rating))

    top_five = sorted(top_five, key=lambda x: x[1], reverse=True)

    top_five = [i for i in sorted(top_five, key=lambda x: x[0])]
    for a in top_five:
        print(a[0])

q1()
print('----------')
q2()
```
